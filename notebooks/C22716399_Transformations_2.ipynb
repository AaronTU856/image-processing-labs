```python
# Get the shape of the image

h, w, d = I.shape

# Scale by factor no cropping
scale_factor = 2.0
S = cv2.resize(I, None, fx=scale_factor, fy=scale_factor, interpolation=cv2.INTER_LINEAR)

# Crop a small section of the image
C = I[0:180, 0:270]

# Find the centre point of the image
cx, cy = w // 2, h //2

# Rotation angle and scale
angle = 45
scale = 1.0

#create rotation matrix
M = cv2.getRotationMatrix2D(center=(cx,cy), angle = angle, scale = scale)
# R = cv2.warpAffine(I, M = M, dsize=(w,h))

# Claculate new bounding dimensions
cos_val = np.abs(M[0, 0])
sin_val = np.abs(M[0, 1])
new_w = int((h * sin_val) + (w * cos_val))
new_h = int((h * cos_val) + (w * sin_val))

# Adjust rotation matrix to fit full screen
M [0, 2] += (new_w / 2) - cx
M[1, 2] += (new_h /2) - cy



# Apply rotation
R = cv2.warpAffine(I, M, (new_w, new_h))

```

```python
# Display Images
plt.figure(figsize=(12,6))

plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(I, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(S, cv2.COLOR_BGR2RGB))
plt.title('Scaled 2x')
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(R, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image 45 deg')
plt.axis("off")

plt.show()
```
