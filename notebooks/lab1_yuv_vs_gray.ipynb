```python
%matplotlib inline

import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Read an image
```

```python
# read in an image
I = cv2.imread("data/samples/Mountains_Scotland.jpg")

if I is None:
    raise Exception("Image not found")

```

```python
# Convert to YUV and extract Y 
```

```python
# convert to YUV
YUV = cv2.cvtColor(I, cv2.COLOR_BGR2YUV)

# Y channel (luminance)
Y = YUV[:,:,0]

```

```python
# convert to grayscale
GRAY = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)
```

```python
# Show wth opencv
```

```python
# show luminance and grayscale with OpenCV
cv2.imshow("Y channel (luminance)", Y)
cv2.imshow("Grayscale (intensity)", GRAY)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Show side by side
```

```python
# side by side plot using matplotlib
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Y channel")
plt.imshow(Y, cmap="gray")
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Grayscale)
plt.imshow(GRAY, cmap='gray')
plt.axis("off")

plt.show()
          
```

```python
# To show the original beside Y channel and Grayscale
```

```python
import cv2
import matplotlib.pyplot as plt

I = cv2.imread("data/samples/Mountains_Scotland.jpg")

YUV = cv2.cvtColor(I, cv2.COLOR_BGR2YUV)
Y = YUV[:,:,0]
GRAY = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(cv2.cvtColor(I, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Y channel")
plt.imshow(Y, cmap="gray")
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Grayscale")
plt.imshow(GRAY, cmap="gray")
plt.axis("off")

plt.show()
```
