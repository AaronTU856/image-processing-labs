```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Open any image
img = cv2.imread('/Users/aaronbaggot/Desktop/image-processing-labs/data/samples/building.jpg', cv2.IMREAD_GRAYSCALE)

# 2. Use Sobel to find horizontal and vertical gradients
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

# Gradient magnitude
magnitude = cv2.magnitude(sobelx, sobely)

# 3. Create edge mask using Canny (255 for edges, 0 for not)
edges = cv2.Canny(img, 100, 200)

# 4. On the edges, show the magnitude of the gradients
edge_magnitude = np.zeros_like(magnitude)
edge_magnitude[edges > 0] = magnitude[edges > 0]

plt.figure(figsize=(12,6))
plt.subplot(1,3,1), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.subplot(1,3,2), plt.imshow(edges, cmap='gray'), plt.title('Canny Edges')
plt.subplot(1,3,3), plt.imshow(edge_magnitude, cmap='inferno'), plt.title('Gradient Magnitude on Edges')
plt.show()

```
